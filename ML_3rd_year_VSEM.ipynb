```python
import numpy as np
from sklearn.linear_model import LinearRegression

# Dataset
X = np.array([2, 3, 5, 7]).reshape(-1, 1)  # Reshaping X to be a 2D array
Y = np.array([4, 5, 7, 10])

# Create a Linear Regression model
model = LinearRegression()

# Fit the model on the dataset
model.fit(X, Y)

# Get the intercept and slope (b0 and b1)
intercept = model.intercept_
slope = model.coef_[0]

# Display the regression equation
print(f"Regression equation: Y = {intercept:.4f} + {slope:.4f}X")

# Predict a value of Y for a given X
X_new = np.array([[4]])  # Predict Y when X = 4 (example)
Y_pred = model.predict(X_new)
print(f"Predicted value of Y for X = 4: {Y_pred[0]:.4f}")
```
